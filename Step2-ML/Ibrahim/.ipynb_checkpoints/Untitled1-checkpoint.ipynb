```python
# Imports
import pandas as pd
import numpy as np
from splinter import Browser
from bs4 import BeautifulSoup
import requests
```

```python
# import data
uni_data = pd.read_csv("API_SQL_UNITID-NAME.csv")
uni_data.head(5)
```

```python
wiki_url = "https://www.wikipedia.org/wiki/"
inst = uni_data['INSTNM'][2]
inst = inst.replace(" ", "_")
inst_url = wiki_url + inst
inst_url
```

```python
inst_list = list(uni_data['INSTNM'])
# inst_list
```

```python
len(uni_data)
```

```python
snippet_list = []

type_list = []
year_est_list = []
students_list = []
nickname_list = []
campus_list = []
```

```python
for index, row in uni_data.iterrows():
    unitid = row['UNITID']
    inst = row['INSTNM']
    inst = inst.replace(" ", "_")
    inst_url = wiki_url + inst
    
    wiki_response = requests.get(inst_url)
    wiki_soup = BeautifulSoup(wiki_response.text, 'html.parser')
    
    try:
            #     # Snippet
        snippet = wiki_soup.find_all('p')[0].text.strip()
        
    except:
        
        snippet = "NA"
        
    snippet_list.append(snippet)
    
    
```

```python
len(snippet_list)
```

```python
motto_list = []
```

```python
for index, row in uni_data.iterrows():
    unitid = row['UNITID']
    inst = row['INSTNM']
    inst = inst.replace(" ", "_")
    inst_url = wiki_url + inst
    
    wiki_response = requests.get(inst_url)
    wiki_soup = BeautifulSoup(wiki_response.text, 'html.parser')
    
    try:
        inst_facts = pd.read_html(inst_url)
        inst_facts = inst_facts[0]
        inst_facts = inst_facts.set_index(0)


        motto = inst_facts.loc['Motto'][1]

        
    except:
        motto =  'N/A'
    
    motto_list.append(motto)
```
